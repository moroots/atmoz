```python
from atmoz.lidar import TOLNet
tolnet = TOLNet()
```

```python
date_start = "2023-07-11"
date_end = "2023-07-13"

# List of instrument group IDs to filter by. We only want GSFC(ID=2), so this list only contains 2.
group = [2]

# Filter for high-resolution files only
product_ID = [4]

tolnet.import_data(
    min_date=date_start, 
    max_date=date_end, 
    instrument_group=group, 
    product_type=product_ID
    )
```

```python
for key in tolnet.data.keys():
    tolnet.tolnet_curtain_plot(tolnet.data[key])
```
