With this notebook, I will show you how to **get lidar data from the NASA TOLNET API** using the `atmoz` package.
-load the TOLNET module


```python
import importlib 
from atmoz import lidar

importlib.reload(lidar)
```

```python
tolnet = lidar.TOLNET()
```

```python
params = {
    "min_date": "2025-05-01", 
    "max_date": "2025-10-01",
    "instrument_group": [7, 8, 2],
    "product_type": [4],
    "processing_type": [1],
    }

tolnet.import_data(**params)
```

```python
for key, item in tolnet.data.items():
    print(key)
    tolnet.plot_curtain(item, title=key)
```
